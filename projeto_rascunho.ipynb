# Analizando os perfis

```python
df = pd.read_csv('State_of_data_2022.csv')
```

```python
len(df.columns)
```

```python
df = df[["('P1_a ', 'Idade')", "('P1_b ', 'Genero')", "('P1_c ', 'Cor/raca/etnia')", "('P1_i_2 ', 'Regiao onde mora')",
"('P1_l ', 'Nivel de Ensino')","('P1_m ', 'Área de Formação')", "('P2_a ', 'Qual sua situação atual de trabalho?')", "('P2_b ', 'Setor')", 
"('P2_c ', 'Numero de Funcionarios')", "('P2_f ', 'Cargo Atual')", "('P2_g ', 'Nivel')", "('P2_h ', 'Faixa salarial')",
"('P2_i ', 'Quanto tempo de experiência na área de dados você tem?')", "('P2_j ', 'Quanto tempo de experiência na área de TI/Engenharia de Software você teve antes de começar a trabalhar na área de dados?')",
"('P2_k ', 'Você está satisfeito na sua empresa atual?')", "('P2_p ', 'Atualmente qual a sua forma de trabalho?')", 
"('P2_q ', 'Qual a forma de trabalho ideal para você?')", "('P2_r ', 'Caso sua empresa decida pelo modelo 100% presencial qual será sua atitude?')",
"('P4_a_1 ', 'Atuacao')", "('P4_e ', 'Entre as linguagens listadas abaixo, qual é a que você mais utiliza no trabalho?')", "('P5_c ', 'Há quanto tempo você busca uma oportunidade na área de dados?')"]]
#Escolhendo as colunas
```

```python
df.head()
```

```python
df_renomeado = df.copy()
colunas = df.columns
colunas_certas = ['idade', 'sexo', 'raca', 'regiao_brasil', 'nivel_ensino', 
                  'area_formacao', 'situacao_trabalho', 'setor', 
                  'numeros_funcionarios', 'cargo_atual', 'nivel', 'faixa_salarial', 
                  'experiencia_com_dados', 'experiencia_antes_dados', 'satisfacao_empresa', 
                  'forma_de_trabalho', 'forma_de_trabalho_ideal','decisao_presencial', 
                  'atuacao',  'linguagem_mais_usada_trabalho',   
                  'tempo_buscando_oportunidade'] #Nome para as colunas

for i in range(len(colunas)):
    df_renomeado = df_renomeado.rename(columns = {colunas[i]:colunas_certas[i]}) #Renomeando colunas
```

```python
df_renomeado['tempo_buscando_oportunidade'].value_counts()
```

```python
df_renomeado['faixa_salarial'] = df_renomeado['faixa_salarial'].replace({'Menos de R$ 1.000/mês' : '<= 1k',
                                             'de R$ 1.001/mês a R$ 2.000/mês'   : '1k-2k',
                                             'de R$ 2.001/mês a R$ 3.000/mês'   : '2k-3k',
                                             'de R$ 3.001/mês a R$ 4.000/mês'   : '3k-4k', 
                                             'de R$ 4.001/mês a R$ 6.000/mês'   : '4k-6k',
                                             'de R$ 6.001/mês a R$ 8.000/mês'   : '6k-8k',
                                             'de R$ 12.001/mês a R$ 16.000/mês' : '12k-16k',
                                             'de R$ 8.001/mês a R$ 12.000/mês'  : '8k-12k',
                                             'de R$ 16.001/mês a R$ 20.000/mês' : '16k-20k',
                                             'de R$ 20.001/mês a R$ 25.000/mês' : '20k-25k',
                                             'de R$ 25.001/mês a R$ 30.000/mês' : '25k-30k',
                                             'de R$ 30.001/mês a R$ 40.000/mês' : '30k-40k',
                                             'Acima de R$ 40.001/mês' : '>= 40k'}) # Arrumando os valores da  faixa_salarial
```

```python
df_renomeado['satisfacao_empresa'] = df_renomeado['satisfacao_empresa'].replace({True:'Sim', False:'Não'})
```

```python
df_renomeado
```

```python
df_renomeado['situacao_trabalho'].unique()
```
